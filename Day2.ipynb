## Day 2
Print the following pattern in the language of your preference.

Input: 4

image.png

```python
n=int(input("Enter Value: "))
if(n%3==0):
    n=n+1
else:
    pass
for i in range(0, n*n):
    for j in range(0, n*n):
        if(i%3==0 or j%3==0):
            print("*",end=" ")
        elif(i==j or i+1==j or j+1==i):
            print("@",end=" ")
        else:
            print(" ",end=" ")
    print("\r")
```
